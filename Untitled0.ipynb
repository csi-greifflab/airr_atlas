```python
from transformers import EsmTokenizer, EsmForMaskedLM
import torch

REPR_LAYERS = [i for i in range(33)]  # Extract embeddings from all layers

model_path = "westlake-repl/SaProt_650M_AF2"
tokenizer = EsmTokenizer.from_pretrained(model_path)
model = EsmForMaskedLM.from_pretrained(model_path)

device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)

seq = "MdEvVpQpLrVyQdYaKv"
tokens = tokenizer.tokenize(seq)
print(tokens)

inputs = tokenizer(seq, return_tensors="pt")
inputs = {k: v.to(device) for k, v in inputs.items()}

with torch.no_grad():
    all_hidden_states = []
    hidden_states = model.base_model(inputs['input_ids'], inputs['attention_mask'])
    all_hidden_states.append(hidden_states)

print(all_hidden_states)

for el in all_hidden_states:
  print(el)
```
